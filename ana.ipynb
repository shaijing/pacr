```python
import orjson
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open("results/20260114_221446/reg.json", "rb") as f:
    adata = orjson.loads(f.read())
with open("results/20260114_172859/or.json", "rb") as f:
    odata = orjson.loads(f.read())
```

```python
ainfo = adata["ainfo"]
oinfo = odata["oinfo"]
```

```python
fig, ax = plt.subplots()
ax.plot(oinfo["train_acc"])
ax.plot(ainfo["train_acc"], "r")
ax.set_xlabel("Epoch")
ax.set_title("Train Acc")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(oinfo["test_acc"])
ax.plot(ainfo["test_acc"], "r")
ax.set_xlabel("Epoch")
ax.set_title("Test Acc")
plt.show()
```
